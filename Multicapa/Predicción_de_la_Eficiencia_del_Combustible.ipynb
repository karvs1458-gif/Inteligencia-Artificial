```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
auto_mpg = fetch_ucirepo(id=9)  #descarga el dataset

# data (as pandas dataframes)
X = auto_mpg.data.features  #obtiene columnas de entrada
y = auto_mpg.data.targets #obtiene columnas de salida
```

```python
import pandas as pd

df = pd.concat([X, y], axis=1).dropna() #une las columnas de entrada y salida en un solo dataframe
```

```python
#volvemos a definir conjuntos X e y
X = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
#divimos todos los datos en dos: entranamiento y de prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, random_state=1
)
```

```python
#nos muestra las dimensiones del conjunto de:
print(X_train.shape) #entrenamiento
print(X_test.shape) #prueba
```

```python
#escalamos los datos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#definimos el modelo de red neuronal
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Definir el modelo
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1)
])
```

```python
#creamos un optimizador con tasa de aprendizaje de 0.001
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje deseada
learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)
```

```python
#se compila el modelo utilizando Adam y la funcion mse
model.compile(
    optimizer=adam_optimizer,
    loss='mean_squared_error',
    metrics=['root_mean_squared_error'],
)
```

```python
#entrenamos el modelo en este caso con 5 epocas
history = model.fit(
    X_train, y_train,
    epochs=5, batch_size=1,
    validation_data=(X_test, y_test)
)
```

```python
import matplotlib.pyplot as plt

# Graficar la funci√≥n de p√©rdida
plt.plot(history.history['loss'], label='P√©rdida de entrenamiento')
plt.plot(history.history['val_loss'], label='P√©rdida de validaci√≥n')
plt.xlabel('√âpocas')
plt.ylabel('P√©rdida')
plt.legend()
plt.title('Funci√≥n de p√©rdida durante el entrenamiento')
plt.show()
```

```python
#evaluamos el modelo en el conjunto de prueba
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=1)
print(f'Test Mean Absolute Error: {test_mae:.2f}')
```

```python
#definimos las predicciones y comparamos con los valores reales
predictions = model.predict(X_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': predictions.flatten()})
print(comparison.head())
```

```python
#otras m√©tricas
from sklearn.metrics import r2_score, mean_squared_error

r2 = r2_score(y_test, predictions)
print(f'R¬≤: {r2}')

mse = mean_squared_error(y_test, predictions)
print(f'MSE: {mse}')
```
